```python
import pandas as pd
from assign_unique_ids_functions import keep_duplicates, remove_duplicates
pd.options.display.max_columns = 999
```

```python
um1 = pd.read_csv('../combine/input/ase-units.csv.gz')
um2 = pd.read_csv('../combine/input/all-sworn-units.csv.gz')
```

```python
um1.columns
```

```python
um2.columns
```

```python
um1.rename(columns={'End.Date': 'End.Date-ase'}, inplace=True)
um1['Start.Date-ase'] = um1['Start.Date']
print(um1.shape)
um1.head()
```

```python
um2.rename(columns={'End.Date': 'End.Date-asu'}, inplace=True)
um2['Start.Date-asu'] = um2['Start.Date']
print(um2.shape)
um2.head()
```

```python
uids = list(set(um1['UID']) & set(um2['UID']))
print(len(set(um1['UID'])))
print(len(set(um2['UID'])))
print(len(uids))
max(um1['Start.Date'].dropna())
```

```python
um1 = um1[(um1['UID'].isin(uids)) & (um1['Start.Date'].notnull())]
um2 = um2[(um2['UID'].isin(uids)) & (um2['Start.Date'].notnull()) & 
          (um2['Start.Date'] <= max(um1['Start.Date'].dropna()))]
```

```python
um1.shape
```

```python
um2.shape
```

```python
um1 = um1.sort_values(['UID','Start.Date-ase', 'End.Date-ase'], na_position='last')
um2 = um2.sort_values(['UID','Start.Date-asu', 'End.Date-asu'], na_position='last')
```

```python
def number_groups(df, group_id):
    df_nums = []
    current_id = 0
    num = 1
    for i in df[group_id]:
        if i != current_id:
            num = 1
            current_id = i
        else:
            num +=1
        df_nums.append(num)
    df.insert(0, 'Number', df_nums)
    return df
```

```python
um1 = number_groups(um1, 'UID')
um2 = number_groups(um2, 'UID')
```

```python
um1.head(10)
```

```python
um2.head(10)
```

```python
um12 = um1.merge(um2, how='outer', on = ['UID', 'Unit', 'Start.Date', 'Number'])
um12.shape
```

```python
problem_uids = um12[(um12['ase-units_ID'].isnull()) | (um12['all-sworn-units_ID'].isnull())]['UID']
```
